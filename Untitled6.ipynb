```python
# âœ… Step 3: Train YOLOv8 Model
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # Use yolov8s.pt or yolov8m.pt for better accuracy
model.train(
    data="/content/Acute-Lymphoblastic-Leukemia-1/data.yaml",  # Adjust folder name if different
    epochs=50,
    imgsz=640,
    batch=8,
    name="hemato_vision_model"
)
```

```python
# Evaluate the model's performance on the validation set
metrics = model.val()
print(metrics)
```

```python
# Run prediction on a test image
results = model.predict("/content/Acute-Lymphoblastic-Leukemia-1/test/images/istockphoto-1492938828-1024x1024_jpg.rf.e53666968ed5df91fbe92065387f3cc4.jpg", save=True)

# Display results
for r in results:
    r.show()  # This will open image with boxes (if using a local environment)
```

```python
# Load the model you just trained
trained_model = YOLO("runs/detect/hemato_vision_model3/weights/best.pt")

# Use it for prediction
results = trained_model.predict("/content/Acute-Lymphoblastic-Leukemia-1/test/images/istockphoto-1492938828-1024x1024_jpg.rf.e53666968ed5df91fbe92065387f3cc4.jpg", save=True)
```

```python
# Print details of predictions
for r in results:
    print(r.boxes)          # Bounding boxes
    print(r.boxes.xyxy)     # Box coordinates (x1, y1, x2, y2)
    print(r.boxes.conf)     # Confidence scores
    print(r.boxes.cls)      # Predicted class IDs
```
